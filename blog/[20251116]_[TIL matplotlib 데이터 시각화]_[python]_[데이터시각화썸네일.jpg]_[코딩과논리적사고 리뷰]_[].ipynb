### 기본 그래프 생성하기
* plt.plot(): 전달된 데이터 리스트로 선 그래프를 그립니다.
* plt.show(): 그래프를 화면에 출력합니다.

```python
# ----------------------------------------------------
# 📈 기본 선 그래프 생성하기 (Matplotlib 실습)
# ----------------------------------------------------

# Matplotlib의 pyplot 모듈 불러오기  #모듈이 모여서 라이브러리가 된다. 이 모듈 안에는 또 여러가지 함수가 있겠지
from matplotlib import pyplot as plt
plt.rc ('font', family='NanumBarunGothic')  #참고로 plt.rc는 맷플롯립이 실행되는 동안 쓰일 그래프 전체의 스타일을 설정하는 함수이다.
# 즉, 그래프를 그릴 때 적용되는 기본 성질을 설정한다.

# 리스트 데이터를 이용해 간단한 선 그래프 그리기
plt.plot([10, 20, 30, 40])  # y축으로 10, 20, 30, 40으로 그래프 그려줘! 이런 뜻임 x축의 값은 자동으로 설정된다.
# [10, 20, 30, 40]는 y축의 값, 자동으로 0, 1, 2, 3... x축의 값
#  결국, plt.plot([0, 1, 2, 3], [10, 20, 30, 40])과 같음 # plt.plot()는 숫자 데이터를 받아서 선 그래프를 그리는 함수이다.

# 그래프 제목 추가
plt.title("기본 선 그래프 예시(자동 x축)")

# x축, y축 이름 지정
plt.xlabel("X축 값")
plt.ylabel("Y축 값")

# 그래프 화면에 표시
plt.show() # x축은 자동으로 0, 1, 2, 3이 지정됩니다.# 여러 그래프가 있을 떄 여기까지가 한 묶임이다~ 이렇게 보면 된다. (경계선)
import numpy as np
# 눈금 위치 + 라벨(글자)도 함께 바꾸기 (교수님 이후 주신 코드 지피티로!)
plt.xticks(np.arange(0, 3.5, 0.5))   # 0~3까지 0.5 간격 np.arrange(시작, 끝값(포함X),증가폭)
plt.yticks(np.arange(0, 50, 10))     # 0~40까지 10 간격

# 눈금 위치 + 라벨(글자)도 함께 바꾸기(교수님 이후 주신 코드 지피티로! 이런 식으로 우리의 니즈에 따라 그래프를 바꿀 수 있는 것이다)
plt.xticks([0, 1, 2, 3], ['A', 'B', 'C', 'D'])
plt.yticks([10, 20, 30, 40], ['Low', 'Mid', 'High', 'Very High'])
```

```python
# ----------------------------------------------------
# 📊 x, y 값으로 선 그래프 그리기 (Matplotlib 실습 2)
# ----------------------------------------------------

from matplotlib import pyplot as plt
plt.rc('font', family='NanumBarunGothic')

# x축 데이터와 y축 데이터 정의
x = [0, 1, 2, 3]
y = [10, 20, 30, 40]  #즉 여기는 눈금을 표시하는게 아니라 x와 y로 점을 찍는거다!

# x, y 값의 대응 관계를 이용해 선 그래프 생성
plt.plot(x, y)
# plt.plot([0, 1, 2, 3], [10, 20, 30, 40])

# 그래프 제목과 축 이름 지정
plt.title("기본 선 그래프 예시 (x, y 데이터 사용)")
plt.xlabel("X축 값")
plt.ylabel("Y축 값")

# 그래프 출력
plt.show()  # (0,10), (1,20), (2,30), (3,40)을 연결한 선 그래프
```

### 제곱 리스트를 이용한 예시 그래프
* 이 예시에서는 x 리스트에 저장된 제곱수를 사용해 그래프를 그립니다.

```python
# ----------------------------------------------------
# 🧮 자연수의 제곱으로 만든 간단한 리스트를 그래프로 그리기
# ----------------------------------------------------

from matplotlib import pyplot as plt
plt.rc('font', family='NanumBarunGothic')

# y값: 1², 2², 3², ..., 8²
y = [1, 4, 9, 16, 25, 36, 49, 64]

# x축 값이 생략되면 자동으로 0, 1, 2, ... 순서로 설정됨
plt.plot([1,2,3,4,5,6,7,8],y)  # plt.plot([0, 1, 2, 3, 4, 5, 6, 7], y) 와 같음

# 그래프 제목 추가
plt.title("자연수의 제곱 그래프 예시")

# 축 이름 지정
plt.xlabel("자연수 (x)")
plt.ylabel("제곱값 (y = x²)")

# 그래프 출력
plt.show()  # (0,1), (1,4), (2,9), ... (7,64) 점들이 선으로 연결됨
```

```python
# ----------------------------------------------------
# 📈 자동 x축 vs 명시적 x축 비교 예시
# ----------------------------------------------------

from matplotlib import pyplot as plt
plt.rc('font', family='NanumBarunGothic')

# 데이터 준비
x = [0, 10, 20, 30, 40, 50, 60, 70]
y = [1, 4, 9, 16, 25, 36, 49, 64]

# 그래프 그리기
plt.plot(x, y)

# 제목 및 축 이름
plt.title("자연수의 제곱 그래프 예시 (x, y 명시)")
plt.xlabel("자연수 (x)")
plt.ylabel("제곱값 (y = x²)")

# 그래프 출력
plt.show()
```

## 🎨 그래프의 색상·모양·마커 지정하기

그래프는 단순히 숫자를 그리는 것뿐 아니라 **시각적 표현을 바꿔서 구분**할 수도 있습니다.

* **색상(color)**: `'b'`(파랑), `'g'`(초록), `'r'`(빨강), `'y'`(노랑), `'k'`(검정)
* **선 스타일(linestyle)**: `'-'`(실선), `'--'`(점선), `':'`(점점선), `'-.'`(혼합선)
* **마커(marker)**: `'o'`(원형), `'s'`(사각형), `'^'`(삼각형), `'x'`(엑스표) 등

💡 **조합 예시**
* `'ro'` → 빨간색(red) 원형(marker='o')
* `'b--'` → 파란색(blue) 점선
* `'g^'` → 초록색(green) 삼각형 마커


```python
# ----------------------------------------------------
# 🎨 그래프 색상·선형·마커 지정 실습
# ----------------------------------------------------

from matplotlib import pyplot as plt
plt.rc('font', family='NanumBarunGothic')

x = [0, 1, 2, 3, 4, 5, 6, 7]
y = [1, 4, 9, 16, 25, 36, 49, 64]

# 1️⃣ 빨간색 실선 그래프
plt.plot(x, y, color='r', linestyle='--', label="빨간색 실선")
plt.legend()   # 👈 범례를 추가해야 label이 화면에 표시됩니다.
plt.show()  #plt.show()를 지정해주니까 여러 그래프를 그려도 경계선이 되어서 구분이 가는구나!! 반대로 plt.show()를 하지 않으면 두 그래프를 한 번에 나타내는 것도 가능하다

# 2️⃣ 파란색 점선 그래프
plt.plot(x, y, 'b:', label="파란색 점선")
plt.legend()
plt.show()

# 3️⃣ 초록색 삼각형 마커 그래프
plt.plot(x, y, 'g^', label="초록색 삼각형 마커")
plt.legend()
plt.show()

#plt.plot(x, y, color='r', linestyle='--', label="빨간색 실선") 과 plt.plot(x, y, 'r--', label="빨간색 실선")는 완전히 같다.
```

## 📊 다양한 그래프 유형 알아보기

* Matplotlib은 선 그래프(line plot) 외에도 여러 형태의 그래프를 지원합니다.
* 데이터의 종류(숫자형, 범주형 등)에 따라 적절한 그래프를 선택합니다.

| 그래프 유형 | 함수 | 주요 용도 |
|--------------|--------|------------|
| 막대 그래프 | `bar()` | 범주별 비교 |
| 히스토그램 | `hist()` | 데이터 분포 확인 |
| 산점도 | `scatter()` | 두 변수 간 관계 |
| 파이 차트 | `pie()` | 구성 비율 표시 |

💡 **Tip:**  
`plt.bar()`, `plt.hist()`, `plt.scatter()`, `plt.pie()` 함수는 모두 `plt.plot()`처럼 비슷한 방식으로 사용됩니다.


## 📊 2️⃣ 막대 그래프(Bar Chart) 그리기

* 막대 그래프는 **범주별(예: 과목, 지역, 제품 등) 값의 크기를 비교**할 때 사용합니다.  
* `plt.bar()` 함수를 사용하면 각 항목의 값을 막대 형태로 시각화할 수 있습니다.

### 주요 옵션
- `plt.bar(x, y)`: x축(범주)과 y축(값)을 지정  
- `color`: 막대 색상 지정 (예: `'skyblue'`, `'orange'`, `'green'`)  
- `plt.title()`, `plt.xlabel()`, `plt.ylabel()`: 제목과 축 이름 설정  

💡 **Tip:**  
막대 그래프는 “어느 항목이 가장 많거나 적은지” 한눈에 비교할 때 가장 직관적입니다.


```python
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt  #위에랑 아래랑 똑같은 것이다. 맷플롯립 안에 있는 pyplot이라는 모듈을 가져와. 그리고 그 이름은 plt로 가져와

plt.rc('font', family='NanumBarunGothic')  # 한글 깨짐 방지

# ----------------------------------------------------
# 📊 막대 그래프 (Bar Chart)
# ----------------------------------------------------

# 데이터 준비
categories = ['국어', '영어', '수학', '과학']
values = [70, 85, 90, 75]

# 막대 그래프 그리기
plt.bar(categories, values, color='skyblue', edgecolor='black')  # edgecolor = 왜곽을 칠해줘! 막대끼리 구분되도록.

# 제목과 축 이름 추가
plt.title('과목별 점수 비교')
plt.xlabel('과목명')
plt.ylabel('점수')
#plt.yticks(np.arange(0,100,10)) 내가해봄. 막대 그래프에도 x와 y의 눈금 표시 시정이 가능하더라!

# 그래프 저장
plt.savefig('/content/drive/MyDrive/mypython/mypandas/과목별_점수_막대그래프.png


# 그래프 표시
plt.show()
```

## 📉 히스토그램(Histogram) 그리기

* 히스토그램은 **데이터가 어떻게 분포되어 있는지**(많이 나오는 구간, 적게 나오는 구간)를 보여줍니다.
* `plt.hist()` 함수는 데이터를 여러 구간(`bins`)으로 나누어 각 구간의 **빈도수**를 막대 그래프로 나타냅니다.

💡 **Tip:**  
데이터의 개수(구간 수)를 바꾸면 그래프 모양이 달라집니다.  
`bins=5` → 큰 구간, `bins=10` → 더 세밀한 구간


```python
from matplotlib import pyplot as plt
plt.rc('font', family='NanumBarunGothic')  # 한글 폰트 설정

# ----------------------------------------------------
# 📉 히스토그램 기본 예제
# ----------------------------------------------------
# 예시 데이터: 학생 10명의 점수
scores = [55, 60, 62, 65, 70, 72, 75, 80, 82, 85]

# 히스토그램 그리기
plt.hist(scores, bins=5, color='skyblue', edgecolor='black')  #bin이 5개라는 것은 히스토그램 통을 5개로 하라는 것이다.즉, 55~85까지를 5통으로 쪼개라!
#여기서 사이에 plt.show() 넣어주면 그래프가 겹치지 않고 두 개가 나온다
plt.hist(scores, bins=10, color='skyblue', edgecolor='black')  #bin 10은 최솟값인 55부터 최댓값 88까지 10개로 쪼개라!

# 한글 제목과 축 이름
plt.title('학생 점수 분포 히스토그램')
plt.xlabel('점수 구간')
plt.ylabel('학생 수')

plt.show()
```

## 🥧 파이 차트(Pie Chart) 그리기

* 파이 차트는 **전체 중에서 각 항목이 차지하는 비율**을 원 모양으로 표현한 그래프입니다.
* `plt.pie()` 함수를 사용해 항목의 비율을 시각화할 수 있습니다.

💡 **주요 옵션**
- `labels`: 각 조각의 이름 (예: 과목명, 지역명 등)
- `autopct='%1.1f%%'`: 비율을 퍼센트로 표시 (소수점 1자리)
- `startangle`: 원이 그려지는 시작 각도 조정


```python
from matplotlib import pyplot as plt
plt.rc('font', family='NanumBarunGothic')  # 한글 깨짐 방지

# ----------------------------------------------------
# 🥧 파이 차트 기본 예제
# ----------------------------------------------------
# 항목과 비율 데이터
label = ['국어', '영어', '수학', '과학']
sizes = [25, 35, 20, 20] #국어는 25% 해당, 영어는 35% 해당... 이런 식인 것이다.

# 파이 차트 그리기
plt.pie(
    sizes,                # 비율 데이터
    labels=label,        # 조각 이름
    autopct='%1.1f%%',    # 퍼센트 표시 (소수점 첫째 자리)
    startangle=140         # 시작 각도(파이 차트가 그려지기 시작하는 각도를 지정: 첫 번째 조각(A에 해당하는 조각)이 140도에서 시작하여 시계 방향으로 배치)
)
#autopct = auto percent의 약자이다. %    1.1f(소숫점 첫째자리까지 표시하는 floating 실수이다)     %% ->이렇게 하면 소숫점 첫째자리까지 파이차트에 나온다.
#인자에 비율 데이터가 먼저 와야한다. 즉, sizes 다음에 labels가 와야한다.  %1.1f%%에서 f 빼먹으면 틀린다.
# 제목 추가
plt.title('과목별 공부 시간 비율')

# 그래프 출력
plt.show()


#“이 차트는 각 과목에 얼마나 공부 시간을 썼는지를 보여줍니다.
# 원 전체가 100%이고, 각 조각은 과목별 비중이에요.
# 예를 들어 영어가 35%라면 전체 공부 시간 중 1/3 이상을 영어에 쓴 겁니다.”
```

## 🎯 산점도(Scatter Plot) 그리기

* 산점도는 **두 변수 간의 관계**를 점(●)으로 나타낸 그래프입니다.  
* `plt.scatter()` 함수는 각 점의 위치를 (x, y) 좌표로 표시합니다.  
  → x값이 커질수록 y값도 함께 커진다면, **양의 상관관계**라고 합니다.  

💡 **Tip:**  
데이터 간의 관계를 한눈에 보고 싶을 때는 막대그래프보다 **산점도**가 더 유용합니다.


```python
from matplotlib import pyplot as plt
plt.rc('font', family='NanumBarunGothic')  # 한글 폰트 설정

# ----------------------------------------------------
# 🎯 산점도 기본 예제
# ----------------------------------------------------
# 간단한 예시 데이터 (x가 커질수록 y도 커지는 관계)
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [2, 4, 5, 7, 8, 10, 11, 14]

# 산점도 그리기
plt.scatter(x, y, color='purple', s=60)  # s=점 크기(size)

# 제목과 축 이름
plt.title('간단한 산점도 예시')
plt.xlabel('공부 시간 (시간)')
plt.ylabel('시험 점수')

plt.show()
```

```python
# 간단한 추세선 추가 (직접 공식 사용)  #이해할 필요 없고 그냥 이걸 쓴다.
plt.scatter(x, y, color='purple', s=60)
plt.plot(x, [2*i for i in x], color='red', label='대략적인 추세선') #이해할 필요 없고 그냥 이걸 쓴다.
plt.title('산점도 + 추세선')
plt.xlabel('공부 시간')
plt.ylabel('시험 점수')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([2, 4, 5, 7, 8, 10, 11, 14])

# 산점도
plt.scatter(x, y, color='purple')

# 📈 회귀선 (추세선)
m, b = np.polyfit(x, y, 1)  # 기울기(m), 절편(b) #x는 독립변수 y는 종속변수 1은 다항식의 차수.
# 즉, 1차 다항식을 만들어라! 이런 뜻  #첫번째가 독립변수 두번째가 종속변수이다.

plt.plot(x, m * x + b, color='red', label='추세선 (Trend Line)')

# 제목 및 축 이름
plt.title('공부시간과 시험점수의 관계')
plt.xlabel('공부 시간')
plt.ylabel('시험 점수')
plt.legend()  #범례를 띄워주는 코드
plt.show()


# “이 빨간 선은 데이터의 전반적인 흐름을 보여주는 추세선(Trend Line) 입니다.
# np.polyfit() 함수가 자동으로 데이터에 가장 잘 맞는 직선을 계산해 줍니다.”

# “나중에 회귀분석(Regression)을 배울 때, 이 개념이 바로 그 기초가 됩니다.”
```
