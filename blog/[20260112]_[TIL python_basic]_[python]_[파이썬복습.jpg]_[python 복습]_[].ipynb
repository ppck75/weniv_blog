```python
# 1) 숫자형의 종류
# int() : 정수  - 괄호 안에 숫자를 넣으면 정수다! 라고 선언된다.
# float() : 실수
# complex(): 복소수 (학문적인 작업에서 많이 쓴다. )


a = int(3)
b = 4
c = float(2)  #2.0이 c에 들어간 것을 알 수 있다.
d = 2.0


# 정수와 실수의 덧셈 뺄셈의 결과는 실수
# 정수와 정수끼리의 결과는 정수이고 ,실수와 실수끼리의 결과는 당연히 실수이다.
e = a + c
f = a - c
g = a * c
h = c / d

#  산술연산자
# a//b : a를 b 로 나눈 몫
4//3
# a % b : a를 b로 나눈 나머지
4%3
# a ** b : a의 b승
4**3
```
