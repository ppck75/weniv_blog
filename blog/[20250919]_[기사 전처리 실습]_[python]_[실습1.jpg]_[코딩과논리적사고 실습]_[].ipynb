이 기사 문자열을 이용해 문자열 관련 함수들을 실습해 봅시다.


```python
article = """ [속보] 삼성전자, 3분기 영업이익 7조 돌파
삼성전자는 이번 분기 실적에서 80조 매출을 기록하며 15% 성장률을 기록했다고 발표했다.
전문가들은 반도체 업황 회복과 함께 Galaxy 스마트폰 판매 호조를 주요 요인으로 꼽았다.
업계에서는 Apple iPhone 16 공개가 글로벌 시장 경쟁을 더욱 치열하게 만들 것이라고 분석했다.
또한 AI 기술 발전이 뉴스 생산과 광고 산업 전반에 큰 변화를 가져올 것으로 예상된다.
한편, LG Display는 OLED TV 신제품을 발표하며 해외 시장 공략을 강화하고 있다.
[관련기사] 애플, 아이폰16 공개 임박 """

print(article)
```

## 문제 2. 단어 검색

## 문제 4. 불필요한 문자 처리

4-1) 기사 문자열 앞뒤에 있는 공백과 줄바꿈을 제거하시오. (`strip`)  
4-2) 기사에서 `"[속보]"` 같은 꼬리표(tag)를 제거하시오. (`replace`)

```python
# 여기에 코드를 작성하세요
# 4-1) article 객체의 앞뒤 공백 제거해서 새로운 문자열 객체를 새로운 변수(예, cleaned_article)에 할당
cleaned = article.strip()

# 4-2) [속보]와 [관련기사] 꼬리표 제거를 제거하면서 각 꼬리표 뒤의 공백도 제거
cleaned = cleaned.replace("[속보] ", "").replace("[관련기사] ", "")

# 4-3) "삼성전자, 3분기 영업이익 7조 돌파"를 "삼성전자, 3분기 영업이익 7조 돌파."으로 대체해서 마침표 보정하기
cleaned = cleaned.replace("삼성전자, 3분기 영업이익 7조 돌파", "삼성전자, 3분기 영업이익 7조 돌파.")

# 4-4) 기사에서 문장 사이의 줄바꿈(\n) 없애기
cleaned = cleaned.replace("\n", "")

# 4-5) 위에서 전처리한 새로운 기사 객체를 출력하기
print(cleaned)
```

## 문제 5. 문장 단위 나누기

5-1) 기사를 `"."` 기준으로 잘라 문장 단위 리스트로 만드시오. (`split`)  
5-2) 나눈 문장 중 첫 번째 문장('삼성전자, 3분기 영업이익 7조 돌파')만 출력하시오. 인덱싱 이용할 것

```python
# 여기에 코드를 작성하세요
# 5-1) split() 함수를 이용해서 기사를 "." 기준으로 잘라 각각의 문장을 요소로 하는 문장 단위 리스트 객체(예, sentences_lst) 만들기
sentences = cleaned.split('.')

# 5-2) 인덱싱을 이용해서 위에서 생성한 리스트의 첫 번째 요소인 '삼성전자, 3분기 영업이익 7조 돌파' 문장을 출력
print(sentences[0])

```
