# 문제 1
* 온라인 서점에서 인기 있는 책 제목들을 크롤링한 결과, '파친코', '지구에서 한아뿐', '아몬드'라는 세개의 책 제목 데이터를 출력했다. 이 데이터를 리스트 형식으로 "books" 변수에 할당하여 출력하세요.

```python
books = ["파친코","지구에서 한아뿐","아몬드"]
print(books)
```

# 문제 3

* 출력값:

# 문제 4
* "books" 리스트에서 모든 책 제목을 ', '로 구분해 하나의 문자열로 합치고 books_str변수에 할당해서 출력하시오.
* 힌트: '구분자'.join() 함수
* 출력값:
* '파친코, 지구에서 한아뿐, 아몬드', 달러구트 꿈 백화점'

```python
books_str = ",".join(books)
print(books_str)
```

# 문제 10

*   뉴스 사이트에서 크롤링한 단어를 두 집합으로 "news_words"와 "online_words"에 할당하고, 두 집합에서 중복되는 단어를 출력하세요.

*  {"경제", "정치", "스포츠", "기술", "사회"}
*  {"기술", "게임", "사회", "문화", "교육"}


```python
news_words = {"경제", "정치", "스포츠", "기술", "사회"}
online_words = {"기술", "게임", "사회", "문화", "교육"}
print(news_words & online_words)
```
