### Tabular Data란 무엇인가?

Tabular Data(테이블형 데이터)란 행과 열로 구조화된 데이터이다.  
즉, 엑셀, csv 등이다!  
*csv 파일이란? comma seperated value 줄여서 csv이다.(콤마로 파일이 구분되어 있다)*

Tabular에서 각 행은 하나의 관측치이며, 각 열은 하나의 변수(속서)을 의미한는 데이터 구조이다.

- Row(행): 테이블형 데이터에서 가로 방향의 데이터 '집합'이다.(하나의 관측치를 관찰할 때마다 하나의 row 즉, 가로가 쭉쭉 쌓인다)  
- Column(열): 테이블에서 세로 방향의 데이터 '집합'이다. 하나의 열은 특정 송성 또는 변수이다(하나의 관촬되는 값들이 가지는 속성을 구분한 것이다)  
- Value(값): 행과 열이 교차하는 지점에 위치하는 '개별' 데이터이다.  
- Index(인덱스): 행을 구별하는 데에 사용되는 고유한 식별자이다.(즉, 행을 구별하는 식별자는 인덱스이고, 열을 구별하는 식별자는 Column Name이다)  
*칼럼명으로 특정 열을 식별해서 접근할 수 있기 때문이다!*

```python
# Tabular Data와 관련된 속성들을 살펴보자

df.shape # (행,열)의 수를 표시한다.  
df.columns # 열의 이름을 나열한다. Index([ ])에 담아서!  
df.values # Tabular Data를 arrary(배열)로 반환한다. (arrary는 Numpy에서 배웠다!)  
df.index # RangeIndex (start = , stop= , step= )을 반환한다. 

# 주의할 점! 함수/메서드가 아니라 ''속성''이다. ()가 없기 때문에 속성!

괄호가 없으면 메서드가 아니라 속성이다.
속성은 객체의 ‘상태’를,
메서드는 객체의 ‘행동’을 나타낸다.
    
```
