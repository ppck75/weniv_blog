# 문제 1
* 다음은 온라인 서점에서 인기 있는 책 제목들을 기반으로 만든 books 리스트입니다.
* books = ['파친코', '달러구트 꿈 백화점', '지구에서 한아뿐', '아몬드']
* books 리스트가 비어 있는지 확인하여, 비어 있으면 "책 목록이 없습니다"를 출력하고, 비어 있지 않으면 "책 목록이 있습니다"를 출력하세요.
* 출력값: 책 목록이 있습니다
* 힌트: if-else문과 len(books)을 이용하세요.

```python
books = ['파친코', '달러구트 꿈 백화점', '지구에서 한아뿐', '아몬드']
if len(books) == 0:
    print('책 목록이 없습니다')
else:
    print('책 목록이 있습니다')
```

# 문제 2

* books 리스트에 "파친코"가 있는지 확인하여, 있으면 "파친코가 목록에 있습니다"를 출력하고, 없으면 "파친코가 목록에 없습니다"를 출력하세요.
* 출력값: 파친코가 목록에 있습니다
* 힌트: if-else문과 in 리스트를 이용하세요.


```python
if "파친코" in books:
    print("파친코가 목록에 있습니다")
else:
    print("파친코가 목록에 없습니다")
```

# 문제 4
* books 리스트에 "해리포터"가 포함될 때까지 "해리포터"를 추가하고 출력하는 코드를 작성하세요.
* 출력값: ['파친코', '달러구트 꿈 백화점', '지구에서 한아뿐', '아몬드', '해리포터']

# 문제 6

* 출력값: 리스트의 각 요소를 줄바꿈하면서 순차적으로 출력하세요.

# 문제 7

* books 리스트에 있는 각 책 제목의 길이를 다음과 같이 순차적으로 출력하세요(줄바꿈 있음)
* 출력값: 예) 파친코의 길이는 3자입니다 달러구트 꿈 백화점의 길이는 10자입니다 ...
* 힌트: for문을 사용하고 출력시 f-string  포매팅과 포매팅안에 각요소를 순차적으로 할당받을 변수와 len(변수)함수를 사용하세요

```python
books = ['파친코', '달러구트 꿈 백화점', '지구에서 한아뿐', '아몬드', '해리포터']
for book_title_len in books:
    a = book_title_len
    b = len(book_title_len)
    print(f"{a}의 길이는 {b}자입니다")
```

# 문제 9
* books 리스트에서 제목에 "꿈"이 포함된 책은 "꿈이 있는 책", 그렇지 않은 책은 "꿈이 없는 책"으로 구분하여 새로운 리스트로 저장하고 출력하세요.
* 출력값: ['꿈이 없는 책', '꿈이 있는 책', '꿈이 없는 책', '꿈이 없는 책', '꿈이 없는 책']
* 힌트: 빈 리스트 생성하고 book_descriptions변수에 할당(book_descriptions = []), for문과 if-else문, if-else문안에 append() 함수 사용할 것

```python
book_descriptions = []
for book in books:
    if "꿈" in book:
        book_descriptions.append("꿈이 있는 책")
    else:
        book_descriptions.append("꿈이 없는 책")
print(book_descriptions)
```

# 문제 10
* 리스트 컴프리헨션을 이용해서 books 리스트에서 제목에 "꿈"이 포함된 책은 "꿈이 있는 책", 그렇지 않은 책은 "꿈이 없는 책"으로 구분하여 새로운 리스트로 저장하고 출력하세요.
* 출력값: ['꿈이 없는 책', '꿈이 있는 책', '꿈이 없는 책', '꿈이 없는 책', '꿈이 없는 책']
* 힌트: book_descriptions = [    if문    else문   for 문+ in books ]

```python
book_descriptions = ["꿈이 있는 책" if "꿈" in book else "꿈이 없는 책" for book in books]
print(book_descriptions)
```
