### for loop

```python
# The loop contiutes to print each item in the list UNTIL the last one, which is Univ.
list=["Kyung","Hee","Univ"]
for s in list:
    print(s)
```

```python
# range(5) generates the sequence of numbers from 0 to 4.
# Then, the loop continues to print each number in the sequence until the last one, which is 4.
for i in range(5):
    print(i)
```

```python
# range(4) generates the sequence of numbers from 0 to 3 for i.
# Then, the loop continues to add i to sum UNTIL the last element of i, which is 3 (0, 0+1, 1+2, 3+3).
sum=0
for i in range(4):
    sum=sum+i
print(sum)
```

### while loop

```python
# print i, while continuing to add 1 to i AS LONG AS it is smaller than 10.
i=0
while i<10:
    i=i+1
    print(i)
```

## Recoding

```python
# load necessary package
import numpy as np # load numpy as np
import pandas as pd # load pandas as pd
```

```python
# loading CSV file stored in google drive
# your own file path should be defined--i.e., where your file is stored in google drive
file_path='/content/drive/MyDrive/data analysis and coding/TESS.csv'
# save data as tess dataframe
tess=pd.read_csv(file_path)
```

### recoding into a new variable

```python
# defining recoding function
def rel_groups(series):
    if series == 3:
        return 100 # assign 100 to “3=Catholic”
    elif series == 1 or series == 2 or series == 4 or series == 9:
        return 200 # assign 200 to “1, 2, 4, 9=Protestant”
    elif series == 10 or series == 11:
        return 300 # assign 300 to “10, 11=Other.Christian”
    elif series == 13:
        return 400 # assign 400 to “13=Atheist”
    elif series == 5 or series == 6 or series == 7 or series == 8 or series == 12:
        return 500 # assign 500 to “5, 6, 7, 8, 12=Misc.Religion”
    elif series == -1:
        return np.nan # np.nan allows you to handle missing values
```

```python
# apply the defined recoding function to XREL1 and save it as religion5
tess['religion5']=tess['XREL1'].apply(rel_groups)
```

```python
# show frequency of religion5 including missing values in ascending order
tess['religion5'].value_counts(dropna=False).sort_index(ascending=True)
```

### recoding a numeric variable into a character variable

```python
 # create a new variable religion5.text from religion5 while copying all the values
tess['religion5.text']=tess['religion5']
```

```python
# recode numeric values into characters
tess['religion5.text']=tess['religion5.text'].replace([100,200,300,400,500],['Catholic','Protestant','Other Christian','Atheist','Misc Religion'])
```

```python
tess['religion5.text'].value_counts(dropna=False).sort_values(ascending=True) # 항목 빈도수로 정렬
```

```python
# crosstab between recoded string variable (religion5.text in row) and original variable (XREL1 in column)
pd.crosstab(tess['religion5.text'],tess['XREL1'])
```

```python
# Create multiple series
series1 = pd.Series(["male","female","female","male","male","male","female"])
series2 = pd.Series([54, 34, 556, 57, 88, 856, 23])
# Combine series into a DataFrame
data = pd.DataFrame({'gender': series1, 'salary': series2}) # column name
print(data)
```

```python
data['salary'].mean() # including 556 and 856 inflates mean value
```

```python
# define the function to recode values above 500 to nan
def recode_salary(value):
    if value > 500:
        return np.nan # recode values above 500 into missing values
    else:
        return value # copy all the other values
```

```python
# apply the function to the 'salary' column
data['salary'] = data['salary'].apply(recode_salary)
```

```python
data['salary'].value_counts(dropna=False).sort_index(ascending=True) # frequency showing missing value as NaN
```

```python
data['salary'].mean() # 556 and 856 are now treated as missing values
```

### saving data WITHOUT missing values

```python
data_filtered=data.dropna() # save data WITHOUT missing values as a new object
```

```python
data_filtered['salary'].value_counts(dropna=False).sort_index(ascending=True) # frequency showing missing value as NaN
```

```python
data_filtered['salary'].mean()
```

### replacing missing values with mean

```python
# you can also replace missing values with the mean of the other values
data['salary'] = data['salary'].replace(np.nan, np.mean(data['salary']))
```

```python
print(data)
```
